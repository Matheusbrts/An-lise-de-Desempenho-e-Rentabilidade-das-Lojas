```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

url1 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_1.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_4.csv"

loja1 = pd.read_csv(url1)
loja2 = pd.read_csv(url2)
loja3 = pd.read_csv(url3)
loja4 = pd.read_csv(url4)

loja1
```

```python
# Concatenando os dados, considerando a mesma estrutura para todos eles 

loja1['Id_Loja'] = 'Loja 1'
loja2['Id_Loja'] = 'Loja 2'
loja3['Id_Loja'] = 'Loja 3'
loja4['Id_Loja'] = 'Loja 4'

df_lojas_concat = pd.concat([loja1,loja2,loja3,loja4], ignore_index=True)
df_lojas_concat
```

```python
# Verificando o tipo dos valores para realizar o tratamento 

df_lojas_concat.dtypes
```

```python
# Arredondando valores com casas apos a virgula 

df_lojas_concat['Frete'] = df_lojas_concat['Frete'].round(2)
df_lojas_concat['Preço'] = df_lojas_concat['Preço'].round(2)
df_lojas_concat
```

```python
# Tratamento de datas e incluindo colunas de Mes,Ano e dia apartadas 

#Converte tipo de dados para datetime
df_lojas_concat['Data da Compra'] = pd.to_datetime(df_lojas_concat['Data da Compra'], format='%d/%m/%Y')

pos = df_lojas_concat.columns.get_loc('Data da Compra') + 1

# insere colunas de data apartadas 
df_lojas_concat['Dia'] = df_lojas_concat['Data da Compra'].dt.day
df_lojas_concat['Mês'] = df_lojas_concat['Data da Compra'].dt.month
df_lojas_concat['Ano'] = df_lojas_concat['Data da Compra'].dt.year
df_lojas_concat
```

```python
df_lojas_concat
```

# 1. Análise do faturamento 



```python
# Qual loja teve o maior faturamento ?
fat_store = df_lojas_concat.groupby('Id_Loja')['Preço'].sum().reset_index(name='Faturamento_total').sort_values('Faturamento_total', ascending=False)
fat_store

display(fat_store.style.format(
    {'Faturamento_total': '{:,.2f}'},
    thousands='.',
    decimal=','
))
```

```python
# Gerando graficos 

ranking_lojas = fat_store
plt.figure(figsize=(10, 6))

ax = sns.barplot(
    data=ranking_lojas,  
    x='Id_Loja', 
    y='Faturamento_total',
    color='skyblue', 
     
)

# Adiona os valores encima dos rotulos
ax.bar_label(
    ax.containers[0],  
    fmt=lambda v: f'R$ {v/1_000_000:.2f}M', # Formata o número
    fontsize=10,      
    color = 'black')

plt.title('Ranking de Faturamento por Loja', fontsize=16)
plt.xlabel('ID da Loja')
plt.ylabel('Faturamento Total (R$)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# Storytelling apos a analise
```

# 2. Vendas por Categoria

```python

cat_vendas = df_lojas_concat.groupby('Categoria do Produto')['Preço'].sum().reset_index(name = 'Faturamento_total').sort_values('Faturamento_total',ascending=False)
cat_vendas
                    
display(cat_vendas.style.format(
    {'Faturamento_total': 'R$ {:.2f}'},  
    thousands='.', 
    decimal=',' 
))
```

```python
# Gerando graficos 

ranking_categorias = cat_vendas
plt.figure(figsize=(15, 6))

ax = sns.barplot(
    data=ranking_categorias,  
    x='Categoria do Produto', 
    y='Faturamento_total',
    color='#ecd079', 
     
)

# Adiona os valores encima dos rotulos
ax.bar_label(
    ax.containers[0],  
    fmt=lambda v: f'R$ {v/1_000_000:.2f}M', # Formata o número
    fontsize=10,      
    color = 'black')

plt.title('Ranking de Faturamento por Categoria', fontsize=16)
plt.xlabel('Produto')
plt.ylabel('Faturamento Total (R$)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

# 3. Média de Avaliação das Lojas


```python
media_avaliacoes = df_lojas_concat.groupby('Id_Loja')['Avaliação da compra'].mean().reset_index(name='media_avaliacoes').sort_values('media_avaliacoes', ascending=False).round(2)
media_avaliacoes
```

```python
plt.figure(figsize=(10, 6))
sns.set_style("whitegrid") 

#Cria um ranking numérico para o eixo Y
media_avaliacoes['rank'] = media_avaliacoes.index

 #Desenha as linhas do grafico
plt.hlines(
    y=media_avaliacoes['rank'],
    xmin=3.0, 
    xmax=media_avaliacoes['media_avaliacoes'],
    color='skyblue',
    alpha=0.8,
    linewidth=3
)

# Desenhar os pontos dp grafico 
plt.scatter(
    media_avaliacoes['media_avaliacoes'],
    media_avaliacoes['rank'],
    color='blue',
    s=200, # Tamanho do ponto
    alpha=1
)

# --- Personalização ---

plt.yticks(
    media_avaliacoes['rank'],
    media_avaliacoes['Id_Loja'],
    fontsize=12
)

for i, row in media_avaliacoes.iterrows():
    plt.text(
        row['media_avaliacoes'] + 0.1, 
        row['rank'],                   
        f'{row["media_avaliacoes"]:.2f}',
        color='black',
        ha='left',
        va='center',
        fontsize=11
    )

plt.title('Ranking de Média de Avaliações por Loja', fontsize=16)
plt.xlabel('Média de Avaliações', fontsize=12)
plt.ylabel('ID da Loja', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.5) # Remove grade vertical
plt.xlim(3.0, 5.2) # Limite X para melhor visualização
plt.gca().invert_yaxis() # Inverte o Y para que a Loja #1 (melhor) fique no topo
plt.tight_layout()
plt.show()
```
